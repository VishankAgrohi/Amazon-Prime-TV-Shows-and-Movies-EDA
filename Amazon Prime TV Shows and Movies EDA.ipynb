## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

import pandas as pd
import numpy as np
import ast

# Load datasets
df1 = pd.read_csv('/content/credits.csv.zip')
df2 = pd.read_csv('/content/titles.csv.zip')

# Merge on 'id'
df = df1.merge(df2, how='inner', on='id')

print(f"Initial shape: {df.shape}")
print("Columns after merge:", df.columns.tolist())

# --- Safely drop duplicates
df = df.drop_duplicates()

# --- Drop columns with >50% missing values
missing_threshold = 0.5
df = df.loc[:, df.isnull().mean() < missing_threshold]

# --- Drop rows where 'title' is missing (safely using .loc)
if 'title' in df.columns:
    df = df.loc[df['title'].notna()]

# --- Convert release_year to integer safely
if 'release_year' in df.columns:
    df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')

# --- Parse 'cast', 'crew', 'genres' safely only if they exist
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except:
        return []

for col in ['cast', 'crew', 'genres']:
    if col in df.columns:
        df[col] = df[col].fillna('[]').apply(safe_literal_eval)

# --- Feature Engineering
if 'cast' in df.columns:
    df['num_cast'] = df['cast'].apply(lambda x: len(x))

if 'crew' in df.columns:
    def get_director(crew_list):
        for member in crew_list:
            if isinstance(member, dict) and member.get('job') == 'Director':
                return member.get('name')
        return np.nan
    df['director'] = df['crew'].apply(get_director)

if 'genres' in df.columns:
    df['genre_names'] = df['genres'].apply(lambda x: [d.get('name') for d in x if isinstance(d, dict)])

# --- Normalize title
df['title'] = df['title'].str.strip().str.lower()

# --- Drop unneeded columns if present
columns_to_drop = [col for col in ['crew', 'genres'] if col in df.columns]
df = df.drop(columns=columns_to_drop)

# --- Final checks
print(f"Final shape: {df.shape}")
print(df.info())
df.head()

````

#### Chart - 8 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the correct path
titles_df = pd.read_csv("/content/titles.csv.zip")

# Select numeric columns and drop rows with missing values
numeric_cols = ['release_year', 'runtime', 'seasons', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
titles_numeric = titles_df[numeric_cols].dropna()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(titles_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()



````

##### 1. Why did you pick the specific chart?

Answer A correlation heatmap was chosen because it's ideal for visualizing relationships between multiple numeric features at once. It helps to quickly identify which variables are positively or negatively correlated and to what degree.

##### 2. What is/are the insight(s) found from the chart?

Answer There is a moderate positive correlation between imdb_votes and imdb_score (movies with higher votes tend to have slightly better scores).

tmdb_popularity and imdb_votes are also positively correlated, suggesting more popular titles receive more votes.

seasons does not strongly correlate with other variables, likely because it only applies to shows, not movies.

#### Chart - 9 - Pair Plot

````python
# Pair Plot visualization code
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the correct path
titles_df = pd.read_csv("/content/titles.csv.zip")

# Select numeric columns and drop rows with missing values
numeric_cols = ['release_year', 'runtime', 'seasons', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
titles_numeric = titles_df[numeric_cols].dropna()



# Pair plot for selected metrics
pairplot_data = titles_numeric[['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']]
sns.pairplot(pairplot_data, diag_kind='hist')
plt.suptitle("Pair Plot of Key Movie Metrics", y=1.02)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer The pair plot is excellent for visualizing pairwise relationships between multiple variables. It helps detect patterns, correlations, and outliers by showing scatter plots for each pair and histograms for distributions.

##### 2. What is/are the insight(s) found from the chart?

Answer A positive trend is visible between imdb_votes and both tmdb_popularity and imdb_score, confirming that popular and well-rated titles attract more votes.

The distributions show that most titles cluster around lower imdb_votes and tmdb_popularity, with a few outliers far to the right.

There’s a weak but noticeable positive correlation between tmdb_score and imdb_score, indicating some consistency between the two platforms' ratings.
